# **Regresión Lineal Multiple**

```python
#librerías
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from turtle import color
from sklearn.linear_model import LinearRegression
```

```python
#Carga de archivos
car = sns.load_dataset(name = "car_crashes")
car.head()
```

## **Regresión**

```python
#Graficamos
sns.scatterplot(x = "alcohol", y = "total", color = "blue", data = car)
sns.scatterplot(x = "speeding", y = "total", color = "green", data = car)
sns.scatterplot(x = "no_previous", y = "total", color = "yellow", data = car)
```

```python
#Declaramos variables dependientes e independientes. Dependiente = total, Independiente = alcohol
var_indep = car[["alcohol", "speeding", "no_previous"]]
var_depen = car[["total"]]
```

```python
#Se define el modelo como la función de regresión lineal
model = LinearRegression()
#Verificamos la función relacionada al modelo
type(model)
```

```python
#Ajustamos el modelo
model.fit(X = var_indep, y = var_depen)
```

```python
#Verificamos los coeficientes obtenidos para el modelo
#(Dos guiones bajos)
model.__dict__
```

**Modelo Matemático**

*y = 1.20984483 + 0.62534291x1 + 0.01978015x2 + 0.81582765x3*

```python
#Evaluamos la eficiencia del modelo obtenido por medio del coeficiente R de determinación
model.score(var_indep, var_depen)
```

```python
#Hacemos las predicciones
predicciones_accidentes = model.predict(X = car[["alcohol", "speeding", "no_previous"]])
predicciones_accidentes
```

```python
car.insert(0, "Predicciones", predicciones_accidentes)
car
```

```python
#Visualizamos entre total real y total predecido
sns.scatterplot(x = "alcohol", y = "total", color = "blue", data = car)
sns.scatterplot(x = "alcohol", y = "Predicciones", color = "green", data = car)
sns.lineplot(x = "alcohol", y = "Predicciones", color = "red", data = car)
```

```python
#Coeficientes de determinación y correlación respectivamente.
coef_deter = model.score(X = var_indep, y = var_depen)
coef_deter
```

```python
#Corroelación
coef_corre = np.sqrt(coef_deter)
coef_corre
```
